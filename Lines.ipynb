```
import pandas as pd
import datetime
import math
import random
import radar
from faker import Faker
```

```
#library to generate for us some data
fake=Faker()
```

```
#function to generate time-series data and a price value
def generateData(n):
  listdata=[]
  start=datetime.datetime(2019,8,1)
  end=datetime.datetime(2019,8,30)
  delta=end-start

  for i in range(n):
    date=radar.random_datetime(start='2019-08-1', stop='2019-08-30').strftime("%Y-%m-%d")
    price=round(random.uniform(900,1000),4)
    listdata.append([date,price])
  data=pd.DataFrame(listdata, columns=['Date','Price'])
  data['Date']=pd.to_datetime(data['Date'],format='%Y-%m-%d')
  data=data.groupby(by='Date').mean()

  return data
```

```
data=generateData(50)
```
