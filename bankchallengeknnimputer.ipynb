```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('Train.csv')
test=pd.read_csv('Test.csv')
```

```python
#we check to see the number of missing values per each dataset
train.isna().sum().sort_values(ascending=False)

```

```python
test.isna().sum().sort_values(ascending=False)
```

```python
"""
well those are a lot of missing values for sure.
We might affect the accuracy if we try imputing columns with a high percentage of missing values 
So lets check the percentage of missing values and act accordingly

"""

def percentageMissing(data):
    sumMissing=data.isna().sum()
    percMissing=sumMissing/len(data)*100
    return percMissing.sort_values(ascending=False)
```

```python
percentageMissing(train)
```

We have columns missing upto 99% of their values this is costly to impute as our values might distort the accuracy of the model

```python
percentageMissing(test)
```

```python
missingDF=pd.DataFrame([percentageMissing(train),percentageMissing(test)])
```

```python
missingDF=missingDF.T
missingDF.columns=["Train","Test"]
```

```python
missingDF
```

Most of the columns in both train and test have an almost identical percentage of missing values. We however have no missing values for age in test while we have a small percentage of missing values in train.

```python
#we create a function to drop columns which are above a certain threshold percentage when it comes to missing values

def dropMiss(data,threshld):
    threshold=len(data)*threshld # any column with more than this is dropped
    return data.dropna(thresh=threshold,inplace=True,axis=1)
    
dropMiss(train,0.4)
dropMiss(test,0.4)
```

```python
percentageMissing(train)
```

As you can see we have eliminated any column that is over 60% null values
